# implementation example: LSTM base predictor on wind 4d

```
dataset_config = RealDatasetConfig(dataset_name="wind", 
                                   dataset_dir="./data/wind_30locs", 
                                   rolling_window=5, 
                                   standardize=True, 
                                   location_num=4, 
                                   repeat_num=5, 
                                   seed=2026)
```

```
predictor_config = LSTMPredictorConfig(save_dir="./data/wind_predictions/lstm_4d/", 
                                       dim_model=64, 
                                       context_window=50, 
                                       predictor_data_ratio=0.5, 
                                       train_valid_ratio=[0.8,0.2], 
                                       train_valid_test_ratio=[0.8, 0.1, 0.1], 
                                       max_epoch=50, batch_size=8, 
                                       learning_rate=0.001, 
                                       early_stop=3)
```
