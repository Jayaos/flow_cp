# implementation example: wind 2d with LOO base predictor

We use the results obtained by LOO base predictor saved in ./data/wind_predictions/

```python
fcp_config = FCPConfig(dataset_dir="./data/wind_predictions/loo_2d/wind_loo-bootstrap_repeat5_dataset.pkl", 
                      dataset_config_dir="./data/wind_predictions/loo_2d/wind_loo-bootstrap_dataset_config.pkl",
                      saving_dir="./results/wind/loo_2d/", 
                      vf_layer_type="concatlinear", 
                      hidden_dims=[32, 32, 32, 32], 
                      activation=torch.nn.Softplus(), 
                      initial_gaussian_distribution_cov_scale=1,
                      null_condition_prob=0.05,
                      batch_size=4,
                      learning_rate=0.0005,
                      max_epoch=50,
                      additional_training_epoch=0,
                      past_resid_as_feature=True,
                      early_stop=4)
```

```python
encoder_config = TransformerEncoderConfig(model_dim=32, 
                                          num_head=2, 
                                          dim_ff=128, 
                                          num_layer=4, 
                                          dropout=0.1, 
                                          context_window=50)
```

```python
run_fcp(fcp_config, encoder_config, device="cuda")
```

```python
eval_config = FCPEvaluationConfig(dataset_dir="./data/wind_predictions/loo_2d/wind_loo-bootstrap_repeat5_dataset.pkl",
                                  dataset_config_dir="./data/wind_predictions/loo_2d/wind_loo-bootstrap_dataset_config.pkl",
                                  model_dir="./results/wind/loo_2d/",
                                  guidance_scale=1.1, 
                                  target_coverage=0.95, 
                                  sampling_num=4096, 
                                  batch_processing=2, # how many batches to split sampling in solving ODE 
                                  atol=1e-05, 
                                  rtol=1e-05)
```

```python
evaluate_fcp(eval_config, device="cuda")
```

```python
summarize_fcp_results(results_dir="./results/wind/loo_2d/",
                      num_folds=5, data="wind", outcome_dim=2, guidance_scale=1.1)
```
